```python
import pandas as pd
import os

# List of files to clean
files = [
    '/content/interactions.csv',
    '/content/products.csv',
    '/content/purchases.csv',
    '/content/reviews.csv',
    '/content/sessions.csv',
    '/content/users.csv'
]

def clean_dataset(file_path):
    print(f'Cleaning {file_path}...')
    df = pd.read_csv(file_path)

    # Basic cleaning: remove exact duplicates
    initial_shape = df.shape
    df = df.drop_duplicates()

    # Example: Simple missing value handling (filling numeric with mean, categorical with mode)
    for col in df.columns:
        if df[col].isnull().any():
            if pd.api.types.is_numeric_dtype(df[col]):
                df[col] = df[col].fillna(df[col].mean())
            else:
                df[col] = df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown')

    cleaned_path = file_path.replace('.csv', '_cleaned.csv')
    df.to_csv(cleaned_path, index=False)
    print(f'Done. Saved to {cleaned_path}. Original rows: {initial_shape[0]}, Cleaned rows: {df.shape[0]}')
    return cleaned_path

# Process all files
cleaned_files = [clean_dataset(f) for f in files]
```

```python
# Preview one of the cleaned datasets
if cleaned_files:
    df_preview = pd.read_csv(cleaned_files[0])
    display(df_preview.head())
```

```python
print('--- Column Information ---')
print(df_preview.info())

print('\n--- Descriptive Statistics ---')
display(df_preview.describe(include='all'))
```

```python
products_df = pd.read_csv('/content/products_cleaned.csv')

# Perform the join on product_id
merged_df = pd.merge(df_preview, products_df, on='product_id', how='inner')

print(f'Joined Data Shape: {merged_df.shape}')
display(merged_df.head())
```

```python
# Fix the warning by specifying a format or handling the partial timestamp correctly
# Note: The 'timestamp' in this dataset appears to be MM:SS.s relative to a session,
# but we will standardize other categorical and date columns across all datasets.

def standardize_df(df):
    # Standardize string columns: lowercase and strip
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].astype(str).str.strip().str.lower()
    return df

# Standardize merged_df
merged_df = standardize_df(merged_df)

# Standardize and structure remaining cleaned datasets
other_datasets = {
    'users': '/content/users_cleaned.csv',
    'purchases': '/content/purchases_cleaned.csv',
    'reviews': '/content/reviews_cleaned.csv',
    'sessions': '/content/sessions_cleaned.csv'
}

structured_dfs = {}
for name, path in other_datasets.items():
    temp_df = pd.read_csv(path)
    temp_df = standardize_df(temp_df)

    # Specific Date conversions where applicable
    date_cols = ['signup_date', 'purchase_date', 'review_date', 'session_start', 'session_end']
    for col in date_cols:
        if col in temp_df.columns:
            temp_df[col] = pd.to_datetime(temp_df[col], errors='coerce')

    temp_df.to_csv(path, index=False)
    structured_dfs[name] = temp_df

print('Standardization and structuring complete for all datasets.')
display(merged_df.head(2))
```

## Execurive Summary of data processing
We have sucessfully transformed the raw datasets into structured, high_quality, and analysis_ready format. Below is the summary of the operations performed:

**Data Cleaning & Integrity**

**Duplicate Removal**: We processed all six datasets (interactions, products, purchases, reviews, sessions, and users) and removed all exact duplicate rows to ensure data uniqueness.

**Missing Value Handling:** We addressed null values by imputing numerical columns with the mean and categorical columns with the most frequent value (mode), ensuring complete records for all 100,000 interaction entries.
Data Integration

**Unified View:** We performed an inner join between the interactions and products datasets. The resulting merged_df combines user behavior with rich product metadata (brand, category, price, and ratings), expanding the feature set from 7 to 18 columns.
Standardization & Normalization

**Text Formatting:** All string-based categorical data (names, categories, brands) has been normalized to lowercase and trimmed of whitespace for consistent grouping and filtering.

**Temporal Structuring:** Identified date fields across all datasets were converted from strings to formal Python datetime objects to enable accurate time-series analysis.
Dataset Profiling

The cleaned interactions data shows approximately 7,000 unique users interacting with nearly 1,000 unique products.
The most common interaction is the 'view' event, with a typical user engagement (dwell time) averaging around 18 seconds.
